## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H=np.matrix([[12,11],
            [7,10]])
J=np.matrix([[0,1,2],
            [7,10,4],
            [3,2,0]])
Hdet=np.linalg.det(H)
Jdet=np.linalg.det(J)

print('---|H|---')
print(Hdet)
print('---|J|---')
print(Jdet)
```

```
HT=np.linalg.inv(H)
JT=np.linalg.inv(J)

print(HT)
print('-------')
print(JT)
```
